# Exploration of the microbiota dataset

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import src.data.dataloader as dataloader
```

## NIPICOL

```python
df_nipicol = dataloader.get_NIPICOL('data')
```

```python
df_nipicol
```

```python
fig, axs = plt.subplots(10, 9, figsize=(20, 30))

i = 0
j = 0

for column in df_nipicol.columns[:-2]:
    df_nipicol.boxplot([column], ax=axs[i][j])
    j += 1
    if j % 9 == 0 and j > 0:
        j = 0
        i += 1
plt.show()
```

```python
fig, axs = plt.subplots(10, 9, figsize=(20, 30))

i = 0
j = 0

for column in df_nipicol.columns[:-2]:
    df_nipicol.plot(x=column, kind='density', ax=axs[i][j])
    j += 1
    if j % 9 == 0 and j > 0:
        j = 0
        i += 1
plt.show()
```
